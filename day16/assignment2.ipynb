```python
# Install required libraries
!pip install -U langchain
!pip install -U langchain-text-splitters

# Import splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Sample text
text = """
Artificial Intelligence is transforming industries across the world.
From healthcare and finance to transportation and education, AI systems
are helping organizations analyze vast amounts of data and automate complex tasks.
Machine learning models can recognize patterns in data that humans may not easily detect.
This capability allows businesses to make better decisions, predict future trends,
and improve efficiency in their operations.

Large Language Models have become one of the most powerful tools in AI.
They can generate text, answer questions, summarize documents, and assist
developers in writing code. These models are trained on massive datasets
and use deep learning architectures such as transformers.
"""

# Create splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50,
    separators=[""]   # FORCE character-level splitting so overlap always appears
)

# Split text
chunks = text_splitter.split_text(text)

print("Total Chunks:", len(chunks))
print("\n")

# Function to extract overlap
def get_overlap(chunk1, chunk2, overlap_size=50):
    end_text = chunk1[-overlap_size:]
    start_text = chunk2[:overlap_size]

    overlap = ""

    for i in range(overlap_size, 0, -1):
        if end_text[-i:] == start_text[:i]:
            overlap = end_text[-i:]
            break

    return overlap


# Display chunks and overlap
for i in range(len(chunks)-1):

    chunk1 = chunks[i]
    chunk2 = chunks[i+1]

    overlap_text = get_overlap(chunk1, chunk2, 50)

    print("=====================================")
    print("Chunk", i+1)
    print(chunk1)

    print("\nChunk", i+2)
    print(chunk2)

    print("\nDetected Overlap:")
    print(overlap_text)

    print("=====================================\n")
```
