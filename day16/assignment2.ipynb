```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

Create a Long Text Document (3 Dense Paragraphs)

```python
text = """
Artificial Intelligence is transforming industries across the world.
From healthcare and finance to transportation and education, AI systems
are helping organizations analyze vast amounts of data and automate complex tasks.
Machine learning models can recognize patterns in data that humans may not easily detect.
This capability allows businesses to make better decisions, predict future trends,
and improve efficiency in their operations.

Large Language Models have become one of the most impactful developments
in modern artificial intelligence. These models are trained on massive datasets
containing books, articles, websites, and other forms of text.
They learn linguistic patterns and relationships between words, allowing them
to generate human-like responses to questions. Applications include chatbots,
virtual assistants, document summarization, and intelligent search systems.

Retrieval-Augmented Generation, commonly known as RAG, combines
information retrieval with generative AI models. Instead of relying solely
on the model’s training data, RAG systems retrieve relevant information from
external documents and provide it as context to the language model.
This improves factual accuracy and ensures responses remain grounded in
real information sources rather than purely generated knowledge.
"""
```

Create the Smart Text Splitter

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50
)
```

Split the Document into Chunks

```python
chunks = text_splitter.split_text(text)

print("Total Chunks Created:", len(chunks))
```

Create Custom Overlap Validation Function

```python
def validate_overlap(chunk1, chunk2, overlap_size=50):

    # Last 50 characters of first chunk
    end_part = chunk1[-overlap_size:]

    # First 50 characters of second chunk
    start_part = chunk2[:overlap_size]

    # Extract actual overlapping characters
    overlap = ""

    for i in range(overlap_size, 0, -1):
        if end_part[-i:] == start_part[:i]:
            overlap = end_part[-i:]
            break

    return overlap
```

Loop Through Chunks and Show Proof

```python
for i in range(len(chunks) - 1):

    chunk1 = chunks[i]
    chunk2 = chunks[i+1]

    overlap_text = validate_overlap(chunk1, chunk2, 50)

    print("\n==============================")
    print(f"Chunk {i+1}:")
    print(chunk1)

    print("\nChunk", i+2, ":")
    print(chunk2)

    print("\nExtracted Overlap:")
    print(overlap_text)

    print("==============================\n")
```
